### Interview grading 

For interview grading today, try correcting the pair sum count code from Tuesday so that it returns the correct answer. Hint: use sets. A helper function is also provided for you (hint)

```python
from typing import List  # this is needed for one of the type hints 

def order_helper(a, b):
    '''
    Returns a and b in order
    '''

    if a < b:
        return (a, b)
    else:
        return (b, a)

def pair_sum_count(numbers: List[int], A: int):
    '''
    Count the unique pairs in numbers that add up to A

    Inputs:
        - numbers: a list of numbers
        - A: a number
    Output:
        - the count of unique pairs in the numbers list that add up to A
    '''
    counter = 0
    seen = set()
    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] + numbers[j] == A:
                pair = order_helper(numbers[i], numbers[j])
                if pair not in seen:
                    counter += 1
                    seen.add(pair)
    return counter
   

pair_sum_count([1,2,3,4,5,6,7], 5)
```

### Defensive programming and errors

- In general, things can go wrong when you program
- It is good to get in the habit of proactively thinking about edge cases
- You can throw errors in your code to catch them
- You are used to thinking of errors as bad. But you can think of them as good. 

```python
def compute_interest(balance: float, rate: float = 1.07):
    # notice the default argument
    return (balance * rate) - balance

compute_interest(100)
```

```python
compute_interest(100, 1.02) 
compute_interest(100, rate=1.02)  # even better, use keyword. More clear
```

What could go wrong?
- Think it over, then demo

```python
def compute_interest(balance: float, rate: float = 1.07):
    # notice the default argument
    if balance < 0:
        raise Exception("Can't compute interest on negative balance")
    return (balance * rate) - balance
```

- Note the syntax above. "Raise"

When would you need exceptions in prior HWs?
- brainstorm this. It is good practice

```python
try:  
    compute_interest(-100, rate=1.02)
except Exception as e:
    print(e)
  
print("All done")
```

### Defensive programming 

- Defensive programming means thinking over what could go wrong and protecting your functions against unexpected but forseeable behavior.
- One lightweight and easy to do this is via assert statements

```python
def compute_interest(balance: float, rate: float = 1.07):
    # notice the default argument
    assert balance > 0, "Can't compute interest on negative balance"
    return (balance * rate) - balance
```
